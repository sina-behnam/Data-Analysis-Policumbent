````python
import pandas as pd
import numpy as np

# read the csv files and create a dataframe
df = pd.read_csv('Diego_15_09_2023_AM_2.csv')

# print the first 5 rows of the dataframe
df.head()
````

````python
# give summary statistics of the dataframe with limited float precision
df.describe()
````

````python
# make plot for each columns in the dataframe which the X axis is the timestamp and the Y axis is the value of the column

import matplotlib.pyplot as plt

# create subplots with 3 columns and 3 rows
fig, axs = plt.subplots(3, 3, figsize=(20, 20))

for column in df.columns:
    if column != 'timestamp':
        index = df.columns.get_loc(column)
        y = np.arange(len(df['timestamp']))
        axs[int(index/3), index%3].plot(y, df[column])
        axs[int(index/3), index%3].set_title(column)

plt.show()
````
